## Demo: How to work with database


Download database file.


```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
query = "SELECT * FROM {}".format(table_name)
df_new = pd.read_sql_query(query, conn)
df_new
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
df_new['ConvertedComp'].plot(kind='hist')
```

### Box Plots


Plot a box plot of `Age.`


```python
plt.figure(figsize=(10, 8))
age_df = df_new['Age'].dropna()
plt.boxplot(age_df, patch_artist=True)
plt.yticks(ticks=range(int(age_df.min()), int(age_df.max()), 5))
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
scat = df_new.dropna()
plt.figure(figsize=(14, 12))
plt.scatter(scat['Age'], scat['WorkWeekHrs'])
plt.grid(True)
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
plt.figure(figsize=(14, 12))
plt.scatter(x=scat['WorkWeekHrs'], y=scat['CodeRevHrs'], s=scat['Age'] ** 1.5, alpha=0.5)
plt.grid(True)
plt.xlabel('Work Week Hours')
plt.ylabel('Code Rev Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
query = "SELECT * FROM DatabaseDesireNextYear"
df_new2 = pd.read_sql_query(query, conn)
pie = df_new2.groupby('DatabaseDesireNextYear')['Respondent'].count().sort_values().tail(5)
plt.figure(figsize=(12, 12))
plt.pie(pie, autopct='%1.1f%%', labels=pie.index)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
age_filtered_df = df_new[(df_new['Age'] >= 30) & (df_new['Age'] <= 35)]
median_workweekhrs = age_filtered_df['WorkWeekHrs'].median()
median_coderevhrs = age_filtered_df['CodeRevHrs'].median()
median_df = pd.DataFrame({
    'WorkWeekHrs': [median_workweekhrs],
    'CodeRevHrs': [median_coderevhrs]
})
plt.figure(figsize=(8, 6))
median_df.plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'])
plt.title('Median Work Week Hours and Code Review Hours for Age Group 30-35')
plt.ylabel('Median Hours')
plt.xticks(ticks=[0], labels=['30-35 Age Group'], rotation=0)
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
age_filtered_df = df_new[(df_new['Age'] >= 45) & (df_new['Age'] <= 60)]
median_convertedcomp = age_filtered_df.groupby('Age')['ConvertedComp'].median()
plt.figure(figsize=(10, 6))
plt.plot(median_convertedcomp.index, median_convertedcomp.values, marker='o', linestyle='-', color='blue')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
mainbranch_counts = df_new['MainBranch'].value_counts()
plt.figure(figsize=(10, 8))
mainbranch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of Main Branch')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.grid(True)
plt.show()
```
